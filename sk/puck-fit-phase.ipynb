# fit

```python
from lmfit import Model, Parameter, report_fit
```

```python
def dphimrd_func(f, a1, a2, a3, a4, a5, a6, fdamp, frd):
    """
    added a6 to modify fdamp
    """
    part1 = a1 + a2*f**(-2) + a3*f**(-1/4.)
    part2 = a4 * (a6*fdamp) / ((a6*fdamp)**2 + (f-a5*frd)**2)
    return part1 + part2
```

```python
def fit(wf):
    model = Model(dphimrd_func, independent_vars=['f'])
    x = wf.fd['freqs']
    y = wf.fd['psi4']['dphase']-wf.fd['psi4']['dphase'][0]

    f1 = wf.fring*0.45
    f2 = wf.fring*1.5
    
    mask = (x>f1) & (x<f2)
    x = x[mask]
    y = y[mask]
    y = y
    
#     plt.figure()
#     plt.scatter(x, y)
#     plt.show()
#     plt.close()
    
    params = model.make_params()
    
    params['fdamp'].value = wf.fdamp
    params['fdamp'].min = wf.fdamp/5
    params['fdamp'].max = wf.fdamp*5
    params['fdamp'].vary=False


    params['frd'].value = wf.fring
#     params['frd'].min = wf.fring-0.1
#     params['frd'].max = wf.fring+0.2
    params['frd'].vary=False

    params['a1'].value = 400
    params['a1'].min = -100000
    params['a1'].max = 100000
    # params['a1'].vary=False

    params['a2'].value = 1
    params['a2'].min = -1000
    params['a2'].max = 1000
    # params['a2'].vary=False

    params['a3'].value = 1
    params['a3'].min = -100000
    params['a3'].max = 100000
    # params['a3'].vary=False

    params['a4'].value = 100
    params['a4'].min = -1000
    params['a4'].max = 1000
    # params['a4'].vary=False

    params['a5'].value = 1
    params['a5'].min = -6
    params['a5'].max = 6
    # params['a5'].vary=False
    
    params['a6'].value = 1
    params['a6'].min = -10
    params['a6'].max = 10
    params['a6'].vary=False

    result = model.fit(y, params, f=x)
    
    report_fit(result.params)
    
    result.plot()
    plt.show()
    plt.close()
    
    return result
```

```python
def cart_to_polar(x, y, z):
    """
    cartesian to spherical polar transformation
    returns: r, theta, phi
    """
    hxy = np.hypot(x, y)
    r = np.hypot(hxy, z)
    theta = np.arctan2(hxy, z)
    phi = np.arctan2(y, x)
    return r, theta, phi
```

```python
psf_coords = []
for wf in wfs:
    r, theta, phi = cart_to_polar(*wf.X1)
#     psf_coords.append([wf.eta, r, theta, phi])
#     psf_coords.append([wf.eta, r])
#     psf_coords.append([wf.eta, theta])
#     psf_coords.append([wf.eta, np.cos(theta)])
    psf_coords.append([wf.eta, np.cos(theta), r])
#     psf_coords.append([wf.eta, phi])
#     psf_coords.append([wf.eta, wf.X1[0]])
#     psf_coords.append([wf.eta, wf.X1[1]])
#     psf_coords.append([wf.eta, wf.X1[2]])
psf_coords = np.array(psf_coords)
```
