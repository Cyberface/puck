# save fit data

```python
def save_data_being_fit(wfs, outdir="fit_data/phase"):
    
    print(f"making output dir: {outdir}")
    os.makedirs(f"{outdir}", exist_ok=True)  
    
    for wf in wfs:
        x = wf.fd['freqs']
        y = wf.fd['psi4']['dphase']-wf.fd['psi4']['dphase'][0]
        
        arr = np.array(list(zip(x, y)))
        
        np.save(os.path.join(outdir, wf.simname + ".npy"), arr)
        
```

```python
from lmfit import Model, Parameter, report_fit
```

```python
def fit(wf):
    model = Model(dphimrd_func, independent_vars=['f'])
    x = wf.fd['freqs']
    y = wf.fd['psi4']['dphase']-wf.fd['psi4']['dphase'][0]

#     f1 = wf.fring_non_prec*0.45
#     f2 = wf.fring_non_prec*1.5

    f1 = wf.fring*0.45
    f2 = wf.fring*1.5

#     f2 = wf.fring_non_prec*1.2
#     f2 = 0.07
#     if 'q8' in wf.simname:
#         f2 = 0.09
    
    mask = (x>f1) & (x<f2)
    x = x[mask]
    y = y[mask]
    y = y
    
#     plt.figure()
#     plt.scatter(x, y)
#     plt.show()
#     plt.close()
    
    params = model.make_params()
    
#     params['fdamp'].value = wf.fdamp_non_prec
#     params['fdamp'].min = wf.fdamp_non_prec/5
#     params['fdamp'].max = wf.fdamp_non_prec*5
    params['fdamp'].value = wf.fdamp
    params['fdamp'].min = wf.fdamp/5
    params['fdamp'].max = wf.fdamp*5
    params['fdamp'].vary=False


#     params['frd'].value = wf.fring_non_prec
    params['frd'].value = wf.fring
#     params['frd'].min = wf.fring_non_prec-0.1
#     params['frd'].max = wf.fring_non_prec+0.2
    params['frd'].vary=False

    params['a1'].value = 400
    params['a1'].min = -100000
    params['a1'].max = 100000
    # params['a1'].vary=False

    params['a2'].value = 1
    params['a2'].min = -1000
    params['a2'].max = 1000
    # params['a2'].vary=False

    params['a3'].value = 100
    params['a3'].min = -100
    params['a3'].max = 100000
    # params['a3'].vary=False

    params['a4'].value = 100
    params['a4'].min = -1000
    params['a4'].max = 1000
    # params['a4'].vary=False

    params['a5'].value = 1
    params['a5'].min = -6
    params['a5'].max = 6
    # params['a5'].vary=False
    
    params['a6'].value = 1
    params['a6'].min = -10
    params['a6'].max = 10
    params['a6'].vary=False


    result = model.fit(y, params, f=x)
    
    report_fit(result.params)
    
    result.plot()
    plt.axvline(wf.fring, c='k', ls='--', label='fRD prediction (prec)')
    plt.axvline(wf.fring_non_prec, c='k', ls='-.', label='fRD prediction (non-prec)')
    plt.legend()
    plt.show()
    plt.close()
    
    return result
```

```python
def cart_to_polar(x, y, z):
    """
    cartesian to spherical polar transformation
    returns: r, theta, phi
    """
    hxy = np.hypot(x, y)
    r = np.hypot(hxy, z)
    theta = np.arctan2(hxy, z)
    phi = np.arctan2(y, x) % (2*np.pi)
    return r, theta, phi
```
