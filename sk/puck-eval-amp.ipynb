# comparison

```python
def amp0(f, eta):
    """
    equation 18: (arxiv:1508.07253)
    """
    return np.sqrt((2 * eta)/(3 * np.pi**(1/3.))) * f**(-7/6.)
```

```python
def compare(wf):
    simname = wf.simname
    
    # get the index for the best-fit coefficients
    psf_idx = psf_names.index(wf.simname)
    
    bf_coords = psf_coords[psf_idx]
    bf_coeffs = psf_y[psf_idx]  

    x = wf.fd['freqs']
    y = wf.fd['strain']['amp'] / amp0(wf.fd['freqs'], wf.eta)

    # find true peak without using first part of data
    # as there can be a peak there sometimes
    mask = x>= 0.05
    true_peak = puck.get_peak_time(x[mask], y[mask])
    
    frd = wf.fring
    dpeak = frd - true_peak
#     print(f"frd: {frd}")
#     print(f"true_peak: {true_peak}")
#     print(f"difference: {dpeak}")
#     print("")
    
#     f_marker = frd
    f_marker = true_peak
    
#     f1 = f_marker/1.15
#     f1 = f_marker/1.05
    f1 = f_marker*0.95
    f2 = f_marker*1.2
    
    mask = (x>f1) & (x<f2)
    x = x[mask]
    y = y[mask]
    y = y
    
    model = Model(ampmrd_func, independent_vars=['f'])
    
    bf_params = model.make_params()
    
    bf_params['fdamp'].value = wf.fdamp
    bf_params['fdamp'].vary=False

    bf_params['frd'].value = wf.fring
    bf_params['frd'].vary=False

    bf_params['gamma1'].value = bf_coeffs[0]
    bf_params['gamma1'].vary=False

    bf_params['gamma2'].value = bf_coeffs[1]
    bf_params['gamma2'].vary=False

    bf_params['gamma3'].value = bf_coeffs[2]
    bf_params['gamma3'].vary=False

    bf_params['gamma4'].value = bf_coeffs[3]
    bf_params['gamma4'].vary=False
    
    y_bf = model.eval(bf_params, f=x)
    
    # model
    parnames = ['gamma1', 'gamma2', 'gamma3', 'gamma4']
    fit_coeffs = dict(fdamp=wf.fdamp, frd=wf.fring)
    
    bf_coords = bf_coords.reshape(1,-1)
    
    for parname in parnames:    
        par_model = psf.load_model(filename=f'amp/{parname}/model.pickle')
        bf_scaled_coords = scale.apply_scaler(bf_coords, amp_X_scalers[parname])
        scaled_yhat = par_model.predict(bf_scaled_coords.reshape(1,-1))
        yhat = scale.apply_inverse_scaler(scaled_yhat.reshape(-1,1), amp_Y_scalers[parname])[:,0]
        fit_coeffs.update({parname:yhat})
    y_model = model.eval(f=x, **fit_coeffs)

    for parname in parnames:    
        par_model_high = psf.load_model(filename=f'amp_high/{parname}/model.pickle')
        bf_scaled_coords = scale.apply_scaler(bf_coords, amp_X_scalers[parname])
        scaled_yhat = par_model_high.predict(bf_scaled_coords.reshape(1,-1))
        yhat = scale.apply_inverse_scaler(scaled_yhat.reshape(-1,1), amp_Y_scalers[parname])[:,0]
        fit_coeffs.update({parname:yhat})
    y_model_high = model.eval(f=x, **fit_coeffs)

    
    
    plt.figure(figsize=(14, 6))
    plt.plot(x, y, label='data')
    plt.plot(x, y_bf, label='best fit', ls='--')
    plt.plot(x, y_model, label='fit')
    plt.plot(x, y_model_high, label='fit (high)', ls='--')
    plt.title(wf.simname)
    plt.legend()
    plt.show()
    plt.close()
```

```python
compare(wfs[0])
```
