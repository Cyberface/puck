```python
wfs = []
for i, f in enumerate(all_files):
    print(f"working: {i}/{len(all_files)}")
    if "q8a08t150dP2_T_96_360" in f:
        continue
    if "q8a04t60D_dPm1" in f:
        continue
    wf = puck.Waveform(filename=f)
    print("info:")
    wf.info
    print("\n")
    
#     wf.compute_fft(t12=[-1400, -1000], t34=[100, 150], plot=True, take_pos_f=True, f_min=0.025, f_max=0.14)
    wf.compute_fft(t12=[-1500, -800], t34=[100, 150], plot=True, take_pos_f=True, f_min=0.018, f_max=0.12)
    
    wfs.append(wf)
```

```python
plt.figure()
for wf in wfs:
    plt.plot(wf.fd['freqs'], wf.fd['psi4']['dphase']-wf.fd['psi4']['dphase'][0])
```

```python
def check_value(value, value_check, dv=0.01):
    if value_check-dv <= value <= value_check+dv:
        return True
    else:
        return False
```

```python
check_value(0.251, 0.25)
```

```python
wfs_q1=[]
wfs_q2=[]
wfs_q4=[]
wfs_q8=[]

for wf in wfs:
    chi = np.sqrt(wf.X1[0]**2+wf.X1[1]**2+wf.X1[2]**2)
    if check_value(chi, 0.8):

        if check_value(wf.eta, phenom.eta_from_q(1)):
            wfs_q1.append(wf)
        elif check_value(wf.eta, phenom.eta_from_q(2)):
            wfs_q2.append(wf)
        elif check_value(wf.eta, phenom.eta_from_q(4)):
            wfs_q4.append(wf)
        elif check_value(wf.eta, phenom.eta_from_q(8)):
            wfs_q8.append(wf)
```

```python
def make_plot(wfs, title, savename=None):
    fig = plt.figure(figsize=(10*2, 4*2))
    for wf in wfs:
        plt.plot(wf.fd['freqs'], wf.fd['psi4']['dphase']-wf.fd['psi4']['dphase'][0], label=wf.simname)
    fig.legend(loc=7)
    fig.tight_layout()
    fig.subplots_adjust(right=0.6)  
    plt.title(title)
    if savename:
        plt.savefig(savename)
```

```python
make_plot(wfs_q1, title="q1", savename='q1.png')
# make_plot(wfs_q2, title="q2", savename='q2.png')
make_plot(wfs_q4, title="q4", savename='q4.png')
make_plot(wfs_q8, title="q8", savename='q8.png')
```
