```python
def get_peak_time(t, amp):
    peak_index = np.argmax(amp)
    peak_time = t[peak_index]
    return peak_time
```

```python
from scipy.interpolate import InterpolatedUnivariateSpline as IUS
def resample(x, y, newx):
    return IUS(x, y)(newx)
```

```python
def setup_compare_non_prec_and_coprec(y_np, y_prec, kind='psi4', t1=None, t2=None):
    
    print("y_np simname {}".format(y_np.simname))
    print("y_prec simname {}".format(y_prec.simname))
    
    
    print("calc initial j frame")
    y_p_coprec = y_prec.__calc_initial_j_frame__()
    
    print("calc coprecessing frame")
    y_p_coprec = y_p_coprec.__calc_coprecessing_frame__()
    
    t0_np = get_peak_time(y_np[2,2][kind].t, y_np[2,2][kind].amp)
    t0_coprec = get_peak_time(y_p_coprec.radiation_axis_info.gwylmo[2,2][kind].t, y_p_coprec.radiation_axis_info.gwylmo[2,2][kind].amp)
    
    t_np_shift = y_np[2,2][kind].t - t0_np
    t_coprec_shift = y_p_coprec.radiation_axis_info.gwylmo[2,2][kind].t - t0_coprec
    
    common_t1 = np.max([t_np_shift[0], t_coprec_shift[0]])
    common_t2 = np.min([t_np_shift[-1], t_coprec_shift[-1]])
    
    print("common t1 = {}".format(common_t1))
    print("common t2 = {}".format(common_t2))
    if t1 is None:
        t1 = common_t1
    if t2 is None:
        t2 = common_t2
    
    newt = np.linspace(t1, t2, 1000)
    
    np_data = {}
    np_data.update({
        't':newt,
        'amp':resample(t_np_shift, y_np[2,2][kind].amp, newt),
        'phi':resample(t_np_shift, y_np[2,2][kind].phi, newt),
        'dphi':resample(t_np_shift, y_np[2,2][kind].dphi, newt),
        'simname':y_np.simname
    })
    
    coprec_data = {}
    coprec_data.update({
        't':newt,
        'amp':resample(t_coprec_shift, y_p_coprec[2,2][kind].amp, newt),
        'phi':resample(t_coprec_shift, y_p_coprec[2,2][kind].phi, newt),
        'dphi':resample(t_coprec_shift, y_p_coprec[2,2][kind].dphi, newt),
        'simname':y_p_coprec.simname
    })
    
    return np_data, coprec_data
```

```python

def td_amp_scale(mtot, distance):
    """
    mtot in solar masses
    distance in m
    M*G/c^2 * M_sun / dist
    """
    return mtot * lal.MRSUN_SI / distance

```

```python
m1, m2 = phenom.m1_m2_M_q(100, 4)
hp, hc = waveform.get_td_waveform(approximant="IMRPhenomD", f_lower=10, delta_t=1./2048/2, mass1=m1, mass2=m2)
```

```python

y_22 = np.abs(lal.SpinWeightedSphericalHarmonic(0, 0, -2, 2, 2))

amp_scale = td_amp_scale(100, 1e6*lal.PC_SI) * y_22

phenD_q4 = {
    'times':phenom.StoM(hp.sample_times.numpy(), 100),
    'hp': hp / amp_scale,
    'hc': hc / amp_scale,
    'amp': waveform.amplitude_from_polarizations(hp, hc) / amp_scale,
    'phase': waveform.phase_from_polarizations(hp, hc)
}


tmp_freq = waveform.frequency_from_polarizations(hp, hc)
tmp_freq = np.concatenate((tmp_freq.numpy(), [0]))
tmp_freq = phenom.HztoMf(tmp_freq, 100) * 2 * np.pi


phenD_q4.update({'freq':tmp_freq})
```
