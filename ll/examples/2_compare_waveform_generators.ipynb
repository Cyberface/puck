## Compare NR calibration data and model using both generators 

```python

# Select single simulation based on name
a = scsearch( catalog=pwca_catalog, keyword='q4a04t90_T_96_360', verbose=True )[0];

# Define data location
datadir = parent(pwca.__path__[0])+'data/version2/'

# Load select waveform data
files = glob( datadir+'q*.txt' )
simnames = [ f_.split('/')[-1].split('.t')[0] for f_ in files ]

# Find index corresponding to chosen case
k = simnames.index( a.simname )

# Load data for this case
raw_data = loadtxt(files[k]).T
data,_,fmin,fmax,fknot = determine_data_fitting_region(raw_data)

# Unpack raw (symmetrised) NR data
raw_f,raw_amp_td,raw_amp_fd,raw_dphi_td,raw_dphi_fd,_,_ = raw_data

# create mask to select positive frequencies
raw_positive_mask = raw_f>0
raw_fp = f_ = raw_f[raw_positive_mask]

# ------------------------------------------------ #
# Generate a coprecessing waveform as 1D array (OLD INPUT STRUCTURE)
# The effective 2-spin mapping is› internal
h22 = generate_pwca_waveform( f_, a.m1, a.m2, a.X1, a.X2, a.L )
# ------------------------------------------------ #

# ------------------------------------------------ #
# External 2-spin mapping
chi_p    = calc_chi_p(a.m1,a.X1,a.m2,a.X2,a.L)
a1,theta = calc_effective_a1_theta( a.m1, a.m2, a.X1, a.X2, a.L )
# Generate a coprecessing waveform as 1D array (NEW INPUT STRUCTURE)
h22b = generate_pwca_waveform_helper( f_, theta, a.eta, a1, chi1, chi2, chi_p )
# ------------------------------------------------ #

#
nr_amp     = raw_amp_fd[raw_positive_mask]

# A
model_amp  = abs(h22)
nr_dphi    = raw_dphi_td[raw_positive_mask]
# -- NOTE -- 
model_dphi = spline_diff( f_, unwrap(angle(h22b)) )

# B
model_amp_b  = abs(h22b)
model_dphi_b = spline_diff( f_, unwrap(angle(h22b)) )

#
calibration_mask    = (f_ >= fmin) & (f_ <= fmax)
adjusted_nr_dphi    = nr_dphi    - min(   nr_dphi[calibration_mask])
adjusted_model_dphi = model_dphi - min(model_dphi[calibration_mask])
adjusted_model_dphi_b = model_dphi_b - min(model_dphi_b[calibration_mask])

#
figure( figsize=1.5*figaspect(0.618/3) )

# Compare model amplitude to NR
# --
subplot(1,2,1)
plot( f_, nr_amp, label='Symmetrised NR' )
plot( f_, model_amp, label='Coprecessing Model (A)' )
plot( f_, model_amp_b, label='Coprecessing Model (B)', ls='--', color='k', alpha=0.5 )
yscale('log')
xscale('log')
ylim( 1e-4, 2e2 )
xlim(0.005,0.2)
axvspan( min(xlim()), fmin, alpha=0.05, color='k')
axvspan( fmax, max(xlim()), alpha=0.05, color='k')
legend()

ylabel(r'$|\tilde{h}_{22}|$')
xlabel('$fM$')
title(a.simname,loc='left',size=12)

# Compare model phase derivative to NR
# --
subplot(1,2,2)
plot( f_, adjusted_nr_dphi, label='Symmetrised NR' )
plot( f_, adjusted_model_dphi, label='Coprecessing Model (A)' )
plot( f_, adjusted_model_dphi_b, label='Coprecessing Model (B)', ls='--', color='k', alpha=0.5 )
ylim(lim(adjusted_model_dphi[calibration_mask],dilate=0.1)+array([0,300]))
xlim(lim([fmin,fmax],dilate=0.5))
axvspan( min(xlim()), fmin, alpha=0.05, color='k')
axvspan( fmax, max(xlim()), alpha=0.05, color='k')
legend()

ylabel(r'$\frac{d}{df}\arg(\tilde{h}_{22})$')
xlabel('$fM$')
title(a.simname,loc='left',size=12)
```

## Directly compare the output of generators A and B

```python
#
figure( figsize=1.5*figaspect(0.618/3) )

subplot(1,2,1)
plot( f_, abs(model_amp-model_amp_b) )
xlabel('$fM$')
title('Difference in Amplitude between Generators A and B')

subplot(1,2,2)
plot( f_, abs(model_dphi-model_dphi_b) )
xlabel('$fM$')
title('Difference in Phase Derivative between Generators A and B');
```
